```python
# Fast, pragmatic geological classification using keyword + embedding similarity
```

```python
# Goal: assign each paper to a coarse geology topic based on Title+Abstract.
```

```python
import re
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Basic cleaning
```

```python
text_series = (df_papers['Title'].fillna('') + ' ' + df_papers['Abstract'].fillna('')).astype(str)
text_clean = text_series.str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
# Define geology categories with seed phrases (lightweight taxonomy)
```

```python
cat_seeds = {
    'Sedimentology & Stratigraphy': [
        'sediment deposition', 'sediment transport', 'grain size', 'stratigraphy', 'facies', 'delta', 'turbidite', 'sandstone', 'shale', 'basin fill'
    ],
    'Marine Geology & Oceanography': [
        'continental shelf', 'deep sea', 'bathymetry', 'seamount', 'abyssal', 'ocean basin', 'pelagic', 'particle flux', 'marine sediments'
    ],
    'Geochemistry & Isotope Geology': [
        'isotope', 'geochemistry', 'trace elements', 'rare earth', 'mineral chemistry', 'stable isotopes', 'radiogenic', 'elemental composition'
    ],
    'Mineralogy & Petrology': [
        'petrography', 'mineralogy', 'igneous', 'metamorphic', 'basalt', 'granite', 'thin section', 'crystallization', 'rock texture'
    ],
    'Economic Geology & Ore Deposits': [
        'hydrothermal', 'mineralization', 'ore deposit', 'metalliferous', 'metal rich', 'sulphide', 'manganese nodule', 'prospecting', 'resource'
    ],
    'Tectonics & Structural Geology': [
        'fault', 'rift', 'fold', 'deformation', 'tectonics', 'subduction', 'orogeny', 'plate boundary', 'fracture zone'
    ],
    'Geomorphology & Surface Processes': [
        'landslide', 'erosion', 'weathering', 'river incision', 'slope stability', 'geomorphology', 'glacial', 'karst', 'dune'
    ],
    'Paleontology & Paleoenvironment': [
        'microfossil', 'fossil', 'paleoenvironment', 'paleoclimate', 'biostratigraphy', 'foraminifera', 'palynology'
    ],
    'Geohazards & Engineering Geology': [
        'earthquake', 'seismic hazard', 'liquefaction', 'tsunami', 'engineering geology', 'groundwater hazard', 'risk assessment'
    ],
    'Hydrogeology': [
        'aquifer', 'groundwater', 'hydrogeology', 'recharge', 'water table', 'contaminant plume'
    ],
    'Geophysics & Remote Sensing': [
        'seismic', 'gravity', 'magnetic', 'geophysical', 'tomography', 'remote sensing', 'satellite', 'inversion'
    ],
    'Climate/Quaternary/Glaciology': [
        'quaternary', 'holocene', 'pleistocene', 'monsoon', 'glacial', 'ice sheet', 'paleoclimate'
    ],
    'Other / Non-geology / Unclear': [
        'n/a'
    ]
}
```

```python
# Build a TF-IDF space over papers + category seed docs
```

```python
seed_docs = []
seed_labels = []
for cat_name, phrases in cat_seeds.items():
    seed_docs.append(' '.join(phrases))
    seed_labels.append(cat_name)

# Define text_clean here to ensure it's available
text_series = (df_papers['Title'].fillna('') + ' ' + df_papers['Abstract'].fillna('')).astype(str)
text_clean = text_series.str.replace(r'\s+', ' ', regex=True).str.strip()

all_docs = pd.concat([text_clean, pd.Series(seed_docs)], ignore_index=True)
```

```python
vectorizer = TfidfVectorizer(
    lowercase=True,
    stop_words='english',
    ngram_range=(1, 2),
    min_df=2,
    max_features=50000
)
```

```python
X_all = vectorizer.fit_transform(all_docs)
```

```python
X_papers = X_all[:len(text_clean)]
X_seeds = X_all[len(text_clean):]
```

```python
sim_mat = cosine_similarity(X_papers, X_seeds)
```

```python
best_seed_idx = sim_mat.argmax(axis=1)
best_score = sim_mat.max(axis=1)
best_cat = pd.Series(best_seed_idx).map(lambda i: seed_labels[int(i)])
```

```python
# Heuristic override rules (keyword triggers) to reduce obvious misclassifications
```

```python
keyword_rules = [
    (re.compile(r'\hydrothermal\|\mineralization\|\metalliferous\|\ore\|\sulphide\|\sulfide\', re.I), 'Economic Geology & Ore Deposits'),
    (re.compile(r'\seamount\|\continental shelf\|\deep[- ]sea\|\abyss\|\ocean basin\|\pelagic\|\bathymetr', re.I), 'Marine Geology & Oceanography'),
    (re.compile(r'\microfossil\|\fossil\|\foraminifera\|\palynolog\|\biostrat', re.I), 'Paleontology & Paleoenvironment'),
    (re.compile(r'\petrograph\|\thin section\|\igneous\|\metamorphic\', re.I), 'Mineralogy & Petrology'),
    (re.compile(r'\seismic\|\gravity\|\magnetic\|\tomograph\|\inversion\', re.I), 'Geophysics & Remote Sensing'),
    (re.compile(r'\fault\|\rift\|\subduction\|\orogen\|\deformation\|\fracture zone\', re.I), 'Tectonics & Structural Geology'),
    (re.compile(r'\groundwater\|\aquifer\|\hydrogeolog\', re.I), 'Hydrogeology'),
    (re.compile(r'\landslide\|\tsunami\|\liquefaction\|\hazard\', re.I), 'Geohazards & Engineering Geology'),
]
```

```python
rule_cat = []
for txt in text_clean.tolist():
    assigned = None
    for pat, cat_name in keyword_rules:
        if pat.search(txt):
            assigned = cat_name
            break
    rule_cat.append(assigned)
```

```python
rule_cat = pd.Series(rule_cat)
final_cat = best_cat.where(rule_cat.isna(), rule_cat)
```

```python
# Confidence buckets (based on similarity)
conf = pd.cut(best_score, bins=[-0.01, 0.08, 0.15, 0.25, 1.0], labels=['low', 'medium', 'high', 'very high'])
```

```python
# Build output dataframe
```

```python
out_cols = ['Title', 'Authors', 'Published Year', 'Journal', 'DOI', 'Study', 'Notes', 'Tags']
keep_cols = [c for c in out_cols if c in df_papers.columns]
```

```python
df_classified = df_papers[keep_cols].copy()
df_classified['Geology Classification'] = final_cat.values
df_classified['Model Similarity Score'] = best_score
# Add short rationale keywords (top contributing tfidf terms) for interpretability
feature_names = np.array(vectorizer.get_feature_names_out())
```

```python
# precompute per-paper top tf-idf terms
```

```python
X = X_papers.tocsr()
top_terms = []
for i_row in range(X.shape[0]):
    row = X.getrow(i_row)
    if row.nnz == 0:
        top_terms.append('')
        continue
    top_idx = row.indices[np.argsort(row.data)[-8:]][::-1]
    terms = feature_names[top_idx]
    top_terms.append(', '.join(terms[:6]))
```

```python
df_classified['Top Terms'] = top_terms
```
